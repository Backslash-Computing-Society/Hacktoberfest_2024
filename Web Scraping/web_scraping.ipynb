### Times Headlines

#### Imports and get page content, define data frames.

```python
from bs4 import BeautifulSoup
import pandas as pd
import requests

r = requests.get('https://timesofindia.indiatimes.com/home/headlines').text
df1=pd.DataFrame(columns=['headline','url'])
df2=pd.DataFrame(columns=['headline','url'])
soup = BeautifulSoup(r, 'html.parser')


all_news = soup.find_all('ul', class_='clearfix')
```

#### Less efficient method using find and find_all to fetch content of tags individually or requiring iteration to find relevant data, a counter has been used to accomodate the problem statement asking for the top 10 results, however that increases the numbetr of if checks and break statements, along with multiple loops for iterations, hence making the task complex.

```python
count=0
max_iterations=10
for ul in all_news:
    list_items = ul.find_all('li')
    for li in list_items:
        spans = li.find_all('span', class_='w_tle')
        for span in spans:
            a_tags=span.find_all('a')
            for a_tag in a_tags:
                if a_tag is not None:
                    headline = a_tag.get('title')
                    url = f"https://timesofindia.indiatimes.com{a_tag.get('href')}"
    
                    print(f"Headline {count+1}\n")
                    print(f'Headline: {headline} \n')
                    print(f'Url: {url} \n')
    
                    df1.loc[len(df1)] = [headline,url]
                    count+=1
    
                    if count >= max_iterations:
                            break
            if count >= max_iterations:
                break
        if count >= max_iterations:
            break
    if count >= max_iterations:
        break
```

### LinkedIn Data Extraction

```python
from bs4 import BeautifulSoup
import requests

url="https://www.linkedin.com/jobs/search/?currentJobId=3957576611&distance=25&geoId=107714860&keywords=software%20engineer%20in%20dehradun&origin=JOB_SEARCH_PAGE_SEARCH_BUTTON&refresh=true"

r=requests.get(url).text
# print(r)
soup=BeautifulSoup(r,"html.parser")

job_list=soup.find('ul',attrs={'class':'jobs-search__results-list'})
list_items=job_list.find_all('li')
id_list=[]
```

### Getting all the job IDs listed in the unordered list by extracting the anchor tags that contain the custom attribute, "data-tracking-control-name" as it contains the job ID as the value for one of its custom attributes. Interestingly, for some of the list items, the said custom attribute is present in div tags rather than anchor tags, this was consolidated by checking for either tag's existence for the said attribute, which succesfully covered the case for all the list items. The IDs were appended into a list.

```python
for item in list_items:
    id_text=item.find('a',attrs={'data-tracking-control-name':'public_jobs_jserp-result_search-card'}).get('data-entity-urn')
    if id_text is not None:
        id_list.append(id_text[-10:])
    else:
        id_div=item.find('div',attrs={'class':'base-card relative w-full hover:no-underline focus:no-underline base-card--link base-search-card base-search-card--link job-search-card'}).get('data-entity-urn')
        if id_div is not None:
            id_list.append(id_div[-10:])
```

```python
print(id_list)
```

#### This code traverses through the extracted list items to now extract the relevant details and print them.

```python

for item in list_items:
    title=item.find("div",attrs={'class':'base-search-card__info'}).h3.text.strip()
    url=item.find("a").get("href")
    company=item.find("div",attrs={'class':'base-search-card__info'}).h4.text.strip()
    location=item.find("div",attrs={"class":"base-search-card__metadata"}).span.text.strip()
    post_time="Posted "+item.find("div",attrs={"class":"base-search-card__metadata"}).time.text.strip()+" on "+item.find("div",attrs={"class":"base-search-card__metadata"}).time['datetime']

    
    print(f"Job Title: {title}")
    print(f"Location: {location}")
    print(f"Company: {company}")
    print(f"Posting Time: {post_time}")
    print(f"URL: {url}")
    print("\n")
```
